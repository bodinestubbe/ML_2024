```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_json("Sarcasm_Headlines_Dataset.json", lines=True)

# Display the first few rows of the dataset
print(data.head())

# Check basic info about the dataset
print(data.info())

# Check for missing values
print(data.isnull().sum())

# Visualize the class distribution
plt.figure(figsize=(8, 6))
sns.countplot(x='is_sarcastic', data=data)
plt.title('Distribution of Sarcastic vs Non-Sarcastic Headlines')
plt.xlabel('is_sarcastic')
plt.ylabel('Count')
plt.show()

# Visualize the length of headlines
data['headline_length'] = data['headline'].apply(len)
plt.figure(figsize=(10, 6))
sns.histplot(data['headline_length'], bins=30, kde=True)
plt.title('Distribution of Headline Length')
plt.xlabel('Headline Length')
plt.ylabel('Count')
plt.show()

# Example of sarcastic and non-sarcastic headlines
print("Example of Sarcastic Headline:")
print(data[data['is_sarcastic'] == 1]['headline'].iloc[0])
print("\nExample of Non-Sarcastic Headline:")
print(data[data['is_sarcastic'] == 0]['headline'].iloc[0])
```
